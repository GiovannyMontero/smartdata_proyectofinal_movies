### Transformacion del archivo "language.csv"

```python
dbutils.widgets.text("param_file_date","2014-12-16")
v_file_date = dbutils.widgets.get("param_file_date")
```

```python
%run "../../utilerias/configurationes"
```

#### Paso 1- Leer el archivo CSV usando "DataFrameReader" de Spark

```python
languages_df = spark.read.table("smartdata_proyect_bronze.languages").filter(f"file_date = '{v_file_date}'") 
```

```python
language_df_select = languages_df.select('languageId', 'languageName')
```

#### Paso 2 - Cambiar el nombre de las columnas

```python
language_df_renamed = language_df_select.withColumnsRenamed({
    "languageId": "language_id",
    "languageName": "language_name"
})
```

```python
from pyspark.sql.functions import current_timestamp, lit
```

#### Paso 4 - Escribir datos en una tabla Delta. 

```python
language_df_renamed.write.mode("overwrite").format("delta").saveAsTable("smartdata_proyect_silver.languages")
```
