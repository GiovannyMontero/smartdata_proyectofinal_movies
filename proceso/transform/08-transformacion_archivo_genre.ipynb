### Transformacion del archivo "genre.csv"

```python
%run "../../utilerias/configurationes"
```

```python
dbutils.widgets.text("param_file_date","2024-12-16")
v_file_date = dbutils.widgets.get("param_file_date")
```

#### Paso 1 - Leer la tabla genre

```python
genres_df = spark.read.table("smartdata_proyect_bronze.genres").filter(f"file_date = '{v_file_date}'")  

```

#### Paso 2 - Seleccionar solo las columnas requeridas

```python
genres_df_selection = genres_df.select("genreId", "genreName","ingestion_date", "enviroment", "file_date")
```

#### Paso 3 - Cambiar el nombre de las columnas

```python
genres_df_renamed = genres_df_selection.withColumnsRenamed({'genreId':'genre_id', 'genreName':'genre_name'})
```

#### Paso 4 - Agregar las columnas requeridas "ingestion_date" y "enviroment"

#### Paso 5 - Escribir los datos en Datalake en formato Parquete

```python
genres_df_renamed.write.mode("overwrite").format("delta").saveAsTable("smartdata_proyect_silver.genres")
```
