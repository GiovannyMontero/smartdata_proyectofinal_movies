### Ingestion del archivo "genre.csv"

```python
%run "../../utilerias/configurationes"
```

```python
dbutils.widgets.text("param_file_date","2024-12-16")
v_file_date = dbutils.widgets.get("param_file_date")
```

```python
dbutils.widgets.text("parm_enviroment","")
v_enviroment = dbutils.widgets.get("parm_enviroment")
```

#### Paso 1 - Leer el archivo genre.csv

```python
# Importamos las librerias
from pyspark.sql.functions import current_timestamp,lit
from pyspark.sql.types import StructType,StructField,IntegerType,StringType
```

```python
schema_genre = StructType([
    StructField('genreId', IntegerType(), False),
    StructField('genreName', StringType(), True)
])
# DBTITLE 1,Leemos el dataset de genres
genres_df = spark.read.option("header",True).schema(schema_genre).csv(f"{raw_folder_path}/{v_file_date}/genre.csv")

```

#### Paso 2 - Agregar las columnas requeridas "ingestion_date" y "enviroment"

```python
genres_df_final = genres_df.withColumn('ingestion_date', current_timestamp()).withColumn('enviroment', lit(v_enviroment)\
).withColumn('file_date', lit(v_file_date))
```

#### Paso 3 - Escribir los datos en Datalake en formato Parquete

```python
genres_df_final.write.mode("overwrite").format("delta").saveAsTable("smartdata_proyect_bronze.genres")
```
