### Configurar Acceso Azure Data Lake Storage Gen2
1. Crear el objeto "External location" llamado "external_location_raw_lakehouse"
2. Crear el objeto "External location" llamado "external_location_bronze_lakehouse"
3. Crear el objeto "External location" llamado "external_location_silver_lakehouse"
4. Crear el objeto "External location" llamado "external_location_gold_lakehouse"

```sql
%sql
--Creacion de un external location en Databricks con leguaje SQL
CREATE EXTERNAL LOCATION IF NOT EXISTS external_location_raw_lakehouse
    URL 'abfss://raw@azmedallonlakehouse.dfs.core.windows.net'
    WITH (STORAGE CREDENTIAL storage_credential_medallon_lakehouse)
    COMMENT 'External location for raw layer'
```
